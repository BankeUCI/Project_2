### Connect to Local Database

```python
rds_connection_string = "postgres:postgres@localhost:5432/Project_2"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

### Load Athletes Table

```python
athletes.to_sql(name='athletes', con=engine, if_exists='append', index=False)
```

### Load Countries Table

```python
countries.to_sql(name='countries', con=engine, if_exists='append', index=False)
```

### Load Events Table

```python
events.to_sql(name='events', con=engine, if_exists='append', index=False)
```
